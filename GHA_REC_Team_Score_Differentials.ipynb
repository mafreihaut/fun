```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sheet_id = "1ed1PxCZtP1U-F4tgbON1jkrRh5E25CcR-QD3damHisY"
sheet_name = "score_diffs"
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"

data = pd.read_csv(url)
data
```

```python
ax1 = data.plot(kind='scatter', x='Tigers', y='Game Week', color='Orange', figsize=(20,4), s=300)  
ax2 = data.plot(kind='scatter', x='Hurricanes', y='Game Week', color='green', ax=ax1, s=300)    
ax3 = data.plot(kind='scatter', x='Hotshots', y='Game Week', color='blue', ax=ax1, s=300)
ax4 = data.plot(kind='scatter', x='Rockets', y='Game Week', color='yellow', ax=ax1, edgecolors='black', s=300) 
ax5 = data.plot(kind='scatter', x='Polar Bears', y='Game Week', color='black', ax=ax1, s=300) 

plt.axvline(x = 2, color = 'r', linestyle = '-')
plt.axvline(x = -2, color = 'r', linestyle = '-')

plt.axvline(x = 2, color = 'r', linestyle = '-')
plt.axvline(x = -2, color = 'r', linestyle = '-')

ax1.set_xlabel("Game Score Differentials")
```
